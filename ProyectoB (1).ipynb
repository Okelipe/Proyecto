```python
#En seguida importamos la base de datos y la tramsformamos en dataframe
import pandas as pd
death=pd.read_csv("DeathData.csv")
df=pd.DataFrame(death)
print(df)
```

```python
#Trabajaremos sobre las medidas de tendencia central
#Media
He=df["HEART"]
media=sum(He)/len(He)
print("La media aritmetica es:",media)
#Usando documentacion de pandas abajo
#print(df.mean())
```

```python
HeO=sorted(He)
HeN=len(He)
index=(HeN-1)//2
if (HeN%2):
    Mediana=HeO[index]
else:
    Mediana=(HeO[index]+HeO[index+1])/2
print("La mediana es",Mediana)
```

```python
He=df["HEART"]
media=sum(He)/len(He)
print("La media aritmetica es:",media)
```

```python
Sum=0
n=len(df["HEART"])
for i in range (n):
    Sum=Sum+(df.loc[i,"HEART"]-media)**2
    #print(i)
    #print(df.loc[i,"HEART"])
    #print (Sum)
N=len(df["HEART"])-1
#print(N)
Var=Sum/N
print("La varianza es",Var)
#print(df.var())
#print(len(df["HEART"]))
```

```python
Desv=Var**(1/2)
print("La desviacion estandar es:",Desv)
```

```python
Rango=He.max()-He.min()
print("El rango es:",Rango)
```

```python
import numpy as np
import math
pd.set_option('precision', 2)
datos = He


# Cálculo del número de intervalos
# Si la parte entera de k es un número impar, redondeamos a la baja
k = 1 + 3.322 * math.log10(len(datos))
numero = int(k)
if numero % 2 == 0:
    periodos = math.ceil(k)
else:
    periodos = int(k)

inf = datos.min()        # Limite inferior del primer intervalo
dif = datos.max()
sup = datos.max() + 1    # Limite superior del último intervalo

intervals = pd.interval_range(
    start=inf,
    end=sup,
    periods=k,
    name="Intervalo",
    closed="left")

df = pd.DataFrame(index=intervals)
df["FreqAbs"] = pd.cut(datos, bins=df.index).value_counts()
df["Marca"]  = df.index.mid
print(df)
```
